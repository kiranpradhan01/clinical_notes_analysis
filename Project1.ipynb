# Setup

```python
import pandas as pd
import glob
import textwrap # get pretty paragraphs
```

```python
# import discharge notes into dataframe
path = r'./DischargeNotes-14/' # use your path
all_files = glob.glob(path + "/*.txt")

li = []

for file in all_files:
    try:
        df = pd.read_csv(file, index_col=None, header=0)
        li.append(df)
    # skip over empty files to avoid parsing error
    except pd.io.common.EmptyDataError:
        print(file," is empty")

notes = pd.concat(li, axis=0, ignore_index=True)
notes
```

```python
# subset discharge notes for "relevant" columns
content = notes[["CATEGORY", "TITLE","TEXT",]]
```
